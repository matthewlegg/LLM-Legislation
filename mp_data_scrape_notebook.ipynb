```python
import requests

member_count = 1

member_name_list = []
member_id_list = []
member_party_list = []
member_json_list = []


for i in range(5000):
  member_id = i+1
  try:
    url = "https://members-api.parliament.uk/api/Members/" + str(i+1)
    res = requests.get(url)
    res_json = res.json()
    membership_status = res_json['value']['latestHouseMembership']['membershipStatus']
    member_house = res_json['value']['latestHouseMembership']['house']
    member_name = res_json['value']['nameDisplayAs']
    member_party = res_json['value']['latestParty']['name']

    if membership_status != None and member_house == 1:
      print('MP number ' + str(member_count) + ': ' + member_name)
      member_name_list.append(member_name)
      member_id_list.append(member_id)
      member_party_list.append(member_party)
      member_json_list.append(res_json)
      member_count +=1
  except:
    print('an error occurred on member id ' +str(member_id))

  if i%50 == 0:
    print('current loop: ' + str(i))
```

```python
member_nameListAs_list = []
member_FullTitle_list = []
member_nameAddressAs_list = []
member_gender_list = []
member_from_list = []

for i in range(650):
  member_nameListAs_list.append(member_json_list[i]['value']['nameListAs'])
  member_FullTitle_list.append(member_json_list[i]['value']['nameFullTitle'])
  member_nameAddressAs_list.append(member_json_list[i]['value']['nameAddressAs'])
  member_gender_list.append(member_json_list[i]['value']['gender'])
  member_from_list.append(member_json_list[i]['value']['latestHouseMembership']['membershipFrom'])

member_json_list[0]['value']
```

```python
print(len(member_name_list))
print(len(member_id_list))
print(len(member_party_list))
print(len(member_json_list))
```

```python
import pandas as pd

# Calling DataFrame constructor after zipping
# all lists, with columns specified
df = pd.DataFrame(list(zip(member_id_list, member_name_list, member_nameListAs_list,
                           member_FullTitle_list, member_nameAddressAs_list, member_from_list,
                           member_party_list, member_gender_list)),
               columns =['member_id', 'display_name', 'list_name', 'full_title', 'address as', 'constituency', 'member_party', 'gender'])
df
```
